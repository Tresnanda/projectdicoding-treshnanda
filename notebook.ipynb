```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

#Import Dataset
melakukan import dataset dan analisis awal terhadap data

```python
# Import dataset yang akan digunakan yaitu bike sharing dataset, dalam kasus ini akan digunakan versi hourly
df = pd.read_csv('hour.csv')
df.head() # Print 5 baris teratas dalam dataframe
```

```python
df.info() # Untuk melihat deskripsi dari dataframe
# Berdasarkan dari output di bawah tidak terdapat kolom yang memiliki nilai null
```

```python
df.describe()
```

```python
categorical_columns = ['season', 'yr', 'mnth', 'hr', 'holiday', 'weekday', 'weathersit', 'workingday']
for col in df.columns:
  if col not in categorical_columns: # Memastikan bahwa nilai yang diprint adalah nilai dari kolom kategorikal
    continue
  print(df[col].value_counts())
```

```python
# Melakukan univariate visualization terhadap kolom kolom kategorikal agar lebih mudah memahami distribusi nilai dalam kolom kategorikal
num_columns = 3
num_rows = (len(categorical_columns) + num_columns - 1) // num_columns

fig, axes = plt.subplots(num_rows, num_columns, figsize=(13, 4 * num_rows))
axes = axes.flatten()

for i, col in enumerate(categorical_columns):
    unique_colors = plt.cm.plasma(np.linspace(0.25, 0.75, len(df[col].value_counts())))
    df[col].value_counts().plot(kind='bar', ax=axes[i], color=unique_colors)
    axes[i].set_title(f'{col}', fontsize=14)
    axes[i].set_ylabel('Jumlah', fontsize=12)
    axes[i].set_xlabel(col, fontsize=12)
    axes[i].tick_params(axis='x', rotation=45)

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

## Faktor faktor apa saja yang mempengaruhi demand sewa sepeda

```python
# Plot jumlah sepeda yang disewa berdasarkan jam dan harinya
plt.figure(figsize=(15, 6))
sns.pointplot(x='hr', y='cnt', hue='weekday', data=df)
plt.title('Demand Sewa Sepeda Berdasarkan Jam dan Hari')
plt.xlabel('Jam')
plt.ylabel('Jumlah Sewa Sepeda')
plt.show()
```

berdasarkan dari plot di atas kita mendapatkan knowledge bahwa pada hari kerja orang orang cenderung untuk menyewa sepeda di pagi hari dan sore hari, sedangkan pada hari libur kerja orang orang cenderung menyewa sepeda pada siang sampai sore hari

```python
# Plot jumlah sepeda yang disewa berdasarkan jam dan harinya
plt.figure(figsize=(15, 6))
sns.pointplot(x='hr', y='cnt', hue='weathersit', data=df)
plt.title('Demand Sewa Sepeda Berdasarkan Jam dan Cuaca')
plt.xlabel('Jam')
plt.ylabel('Jumlah Sewa Sepeda')
plt.show()
```

dari plot di atas kita dapat mengetahui bahwa pada cuaca 1 (cerah) terdapat jumlah paling banyak user menyewa sepeda dan disusul oleh cuaca 2 (sedikit berkabut). sedangkan cuaca 3 (bersalju ringan) mengalami penurunan dalam jumlah sewa dan pada cuaca 4 (cuaca ekstrim) hanya terdapat dua penyewaan sepeda.

```python
# Plot jumlah sepeda yang disewa berdasarkan season / musim
plt.figure(figsize=(15, 6))
sns.pointplot(x='hr', y='cnt', hue='season', data=df)
plt.title('Demand Sewa Sepeda Berdasarkan Jam dan Musim')
plt.xlabel('Jam')
plt.ylabel('Jumlah Sewa Sepeda')
plt.show()
```

dari plot di atas pada musim 2 (musim panas), 3 (gugur), dan 4 (dingin) terdapat jumlah penyewaan yang hampir mirip di luar jam 17 sampai 23. sedangkan pada musim 1 (semi) terdapat penurunan yang signifikan dalam penyewaan sepeda

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='temp', y='cnt', hue='season', data=df)
plt.title('Temperature vs. Total Count (Season Hue)')
plt.xlabel('Temperature')
plt.ylabel('Total Count')
plt.show()
```

dari plot ini kita dapat mengetahui bahwa terdapat korelasi dari suhu cuaca dengan jumlah sepeda yang dirental, yaitu pada cuaca yang suhunya lebih hangat maka lebih banyak sepeda yang tersewa


### Faktor-faktor yang mempengaruhi demand rental sepeda

* **Waktu pada Hari:** Demand memuncak pada jam-jam sibuk pagi dan sore hari, terutama pada hari kerja. Selama akhir pekan dan hari libur, permintaan cenderung lebih tinggi pada sore dan malam hari.
* **Cuaca:** Cuaca yang cerah secara signifikan meningkatkan demand, dengan cuaca yang sedikit berkabut juga relatif menguntungkan.  Kondisi cuaca buruk seperti salju atau cuaca ekstrem secara drastis mengurangi penyewaan sepeda.
* **Musim:** Musim panas, musim gugur, dan musim dingin menunjukkan pola permintaan yang relatif sama, sementara musim semi mengalami penurunan yang cukup besar dalam hal penyewaan.
* **Suhu:** Terdapat korelasi positif yang kuat antara suhu dan jumlah sepeda yang disewa. Suhu yang lebih hangat menyebabkan permintaan yang lebih tinggi.
